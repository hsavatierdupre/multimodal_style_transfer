```python
from skimage.io import imread
from skimage.color import rgb2gray, gray2rgb
import matplotlib.pyplot as plt
import numpy as np
```

```python
cornell_impronte = imread('cornell_impronte/cornell_impronte_with_style_image.jpg')
cornell_impronte_gray = rgb2gray(cornell_impronte)
cornell_impronte_90 = imread('cornell_impronte/cornell_impronte90_with_style_image.jpg')
cornell_impronte_90_gray = rgb2gray(cornell_impronte_90)
diff = abs(cornell_impronte_gray-cornell_impronte_90_gray)
fig = plt.figure()
plt.axis("off")
im = plt.imshow(diff, cmap = 'gray')
plt.colorbar(im)
```

```python
cornell_brushstrokes = imread('cornell_brushstrokes/cornell_brushstrokes_with_style_image.jpg')
cornell_brushstrokes_gray = rgb2gray(cornell_brushstrokes)
cornell_brushstrokes_90 = imread('cornell_brushstrokes/cornell_brushstrokes90_with_style_image.jpg')
cornell_brushstrokes_90_gray = rgb2gray(cornell_brushstrokes_90)
fig = plt.figure()
plt.axis("off")
im = plt.imshow(abs(cornell_brushstrokes_gray-cornell_brushstrokes_90_gray), cmap = 'gray')
plt.colorbar(im)
```

```python
from numpy.random import normal
content = imread('cornell.jpg')
content_gray = rgb2gray(content)
mean = 0
var = 0.005
sigma = var**0.5
content_noisy_gray = content_gray + normal(loc = mean, scale = sigma, size = content_gray.shape)
content_noisy = gray2rgb(content_noisy_gray)
fig = plt.figure()
plt.axis("off")
plt.imshow(content_noisy)
```
